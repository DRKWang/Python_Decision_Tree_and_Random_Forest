```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
%matplotlib inline
```

## for hold data

```python
path = "/Users/binshuaiwang/PycharmProjects/XAI_project/XAI_project/data/HOLD"

airport_location = (40.77157390396803, -73.86862270404231)
# Change the directory 
os.chdir(path) 
fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(airport_location[0], airport_location[1],  marker="*", c="red", s = 200,cmap='rainbow')
count = 0
# iterate through all file 
for file_name in os.listdir(): 
    count += 1
    if count >= 10:
        break
    # Check whether file is in text format or not 
    if file_name.endswith(".dat"): 
        lat, long = [], []
        color = np.random.random(3)
        with open(file_name, "r") as file:
            for line in file:
                line_split = line.split(",")
                lat.append(float(line_split[2]))
                long.append(float(line_split[3]))
        plt.scatter(lat, long,  c=np.array([color]), s = 0.2,cmap='rainbow')

#plt.xlim(0-bound*0.2,100+bound*0.2)
#plt.ylim(0-bound*0.2,100+bound*0.2)
plt.show()
```

## for nohold data

```python
path = "/Users/binshuaiwang/PycharmProjects/XAI_project/XAI_project/data/nohold"

airport_location = (40.77157390396803, -73.86862270404231)
# Change the directory 
os.chdir(path) 
fig, ax = plt.subplots(figsize=(20,20))
plt.scatter(airport_location[0], airport_location[1],  marker="*", c="red", s = 200,cmap='rainbow')
count = 0
# iterate through all file 
for file_name in os.listdir(): 
    count += 1
    if count >= 30:
        break
    # Check whether file is in text format or not 
    if file_name.endswith(".dat"): 
        lat, long = [], []
        color = np.random.random(3)
        with open(file_name, "r") as file:
            for line in file:
                line_split = line.split(",")
                lat.append(float(line_split[2]))
                long.append(float(line_split[3]))
        plt.scatter(lat, long,  c=np.array([color]), s = 0.2,cmap='rainbow',label=count)
#plt.xlim(0-bound*0.2,100+bound*0.2)
#plt.ylim(0-bound*0.2,100+bound*0.2)
plt.legend()
plt.show()
```
